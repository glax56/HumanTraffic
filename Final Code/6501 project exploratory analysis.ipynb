```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('project1.tsv',sep='\t')
```

```python
df.head()
```

```python
df.describe()
```

```python
lst_columns = list(df.columns)
lst_columns
```

```python
for i in lst_columns:
    print(i)
    print(df[i].unique())
```

```python
#fill nans with 0s
df = df.fillna(0)
```

```python
for i in lst_columns:
    print(i)
    print(df[i].unique())
```

```python
cit_count = df.groupby(by='citizenship').count().reset_index()
cit_count = cit_count.sort_values('yearOfRegistration',ascending=False)
top_cit_count = list(cit_count[1:6]['citizenship'])
bot_cit_count = list(cit_count[-110:]['citizenship'])
df_top_cit = df[df['citizenship'].isin(top_cit_count)]
df_bot_cit = df[df['citizenship'].isin(bot_cit_count)]
top_age_breakdown = df_top_cit.groupby(by='ageBroad').count().reset_index()[['ageBroad','yearOfRegistration']].rename(columns={'yearOfRegistration':'count'})
print(top_age_breakdown)
bot_age_breakdown = df_bot_cit.groupby(by='ageBroad').count().reset_index()[['ageBroad','yearOfRegistration']].rename(columns={'yearOfRegistration':'count'})
print(bot_age_breakdown)
```

```python
cit_count
```

```python
top_age_breakdown = top_age_breakdown[top_age_breakdown['ageBroad'] != 0].set_index('ageBroad')
top_age_breakdown.plot.bar(title='Top Countries')
bot_age_breakdown = bot_age_breakdown[bot_age_breakdown['ageBroad'] != 0].set_index('ageBroad')
bot_age_breakdown.plot.bar(title='Bottom Countries')
```

```python
top_gen_breakdown = df_top_cit.groupby(by='gender').count().reset_index()[['gender','yearOfRegistration']].rename(columns={'yearOfRegistration':'count'})
print(top_gen_breakdown)
bot_gen_breakdown = df_bot_cit.groupby(by='gender').count().reset_index()[['gender','yearOfRegistration']].rename(columns={'yearOfRegistration':'count'})
print(bot_gen_breakdown)
```

```python
top_gen_breakdown = top_gen_breakdown[top_gen_breakdown['gender'] != 0].set_index('gender')
top_gen_breakdown.plot.bar(title='Top Countries')
bot_gen_breakdown = bot_gen_breakdown[bot_gen_breakdown['gender'] != 0].set_index('gender')
bot_gen_breakdown.plot.bar(title='Bottom Countries')
```

```python
top_length_breakdown = df_top_cit.groupby(by='traffickMonths').count().reset_index()[['traffickMonths','yearOfRegistration']].rename(columns={'yearOfRegistration':'count'})
print(top_length_breakdown)
bot_length_breakdown = df_bot_cit.groupby(by='traffickMonths').count().reset_index()[['traffickMonths','yearOfRegistration']].rename(columns={'yearOfRegistration':'count'})
print(bot_length_breakdown)
```

```python
top_length_breakdown = top_length_breakdown[top_length_breakdown['traffickMonths'] != 0].set_index('traffickMonths')
top_length_breakdown.plot.bar(title='Top Countries')
bot_length_breakdown = bot_length_breakdown[bot_length_breakdown['traffickMonths'] != 0].set_index('traffickMonths')
bot_length_breakdown.plot.bar(title='Bottom Countries')
```
